```python
import torch
from transformers import AutoModel, AutoProcessor

# Tải mô hình và bộ xử lý
model = AutoModel.from_pretrained("jadechoghari/vfusion3d", trust_remote_code=True)
processor = AutoProcessor.from_pretrained("jadechoghari/vfusion3d")


```

```python
from PIL import Image

# Đường dẫn đến ảnh trong thư mục của bạn
image_path = '/content/anhtest.jpg'  # Thay 'path/to/your/image.jpg' bằng đường dẫn thực tế tới ảnh của bạn

# Mở ảnh từ thư mục
try:
    image = Image.open(image_path)
    print(f"Đã mở ảnh: {image_path}")
except FileNotFoundError:
    print(f"Không tìm thấy ảnh tại: {image_path}")
except Exception as e:
    print(f"Lỗi: {e}")

# Tiền xử lý hình ảnh và lấy thông tin camera nguồn
try:
    image, source_camera = processor(image)
    print("Tiền xử lý ảnh thành công!")
except Exception as e:
    print(f"Lỗi khi tiền xử lý ảnh: {e}")
```

```python
# Tạo các mặt phẳng 3D
output_planes = model(image, source_camera)
print("Kích thước của các mặt phẳng:", output_planes.shape)
```

```python
# Tạo lưới 3D và lưu vào tệp
output_planes, mesh_path = model(image, source_camera, export_mesh=True)
print("Kích thước của các mặt phẳng:", output_planes.shape)
print("Lưới 3D được lưu tại:", mesh_path)
```

```python
# Tạo video và lưu vào tệp
output_planes, video_path = model(image, source_camera, export_video=True)
print("Kích thước của các mặt phẳng:", output_planes.shape)
print("Video được lưu tại:", video_path)
```
